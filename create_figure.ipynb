```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

plt.rc("font", family="D2Coding")

x1 = ["VRIN", "TRIN", "trin(T/F)", "F", "F(B)", "F(P)", "FBS", "L", "K", "S"]
x2 = ["Hs", "D", "Hy", "Pd", "Mf", "Pa", "Pt", "Sc", "Ma", "Si"]
print(x1 + x2)
y1 = [61, 61, "T", 58, 55, 52, 53, 40, 50, 62]
y2 = [61, 58, 59, 48, 49, 55, 56, 61, 62, 65]

x = np.array(x1 + [""] + x2)
y = np.array(y1 + [0] + y2)

data = pd.DataFrame()
data["x"] = x
data["y"] = y

index = data[data["y"].apply(lambda x: isinstance(x, str))].index
droppedData = data.drop(index=index, axis=0)
print(droppedData)

plt.figure(figsize=(10, 6))
plt.plot(
    data=data
)
plt.ylim(20, 120)
plt.yticks(np.arange(20, 120, 10))
plt.title("MMPI_profile_input MMPI-2 검사결과")
plt.grid(True, axis="y", alpha=0.3)

plt.show()
```
