```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

plt.rc("font", family="D2Coding")

x1 = ["VRIN", "TRIN", "trin(T/F)", "F", "F(B)", "F(P)", "FBS", "L", "K", "S"]
x2 = ["Hs", "D", "Hy", "Pd", "Mf", "Pa", "Pt", "Sc", "Ma", "Si"]
# print(x1 + x2)
y1 = [61, 61, "T", 58, 55, 52, 53, 40, 50, 62]
y2 = [61, 58, 59, 48, 49, 55, 56, 61, 62, 65]

x = x1 + [""] + x2
y = y1 + [None] + y2

data = pd.DataFrame()
data["x"] = x
data["y"] = y
# print(data)

index = data[data["y"].apply(lambda y: isinstance(y, str))].index
# print(index)
droppedData = data.drop(index=index)
# print(droppedData)

plt.figure(figsize=(10, 6))
figure = plt.plot(
    droppedData["x"], droppedData["y"], color="blue"
)

plt.ylim(20, 120)
plt.yticks(np.arange(20, 120+1, 10))
plt.title("MMPI_profile_input MMPI-2 검사결과")
plt.grid(True, axis="y", alpha=0.3)
plt.axhline(y=65, color="black", linewidth=0.5)
plt.axvline(x="", color="black", linewidth=0.5, linestyle="--")

for i, txt in enumerate(droppedData["y"]):
    if pd.isna(txt):
        continue
    
    plt.text(
        x=droppedData["x"].iloc[i], 
        y=droppedData["y"].iloc[i]+2,           # 선에 텍스트가 겹치지 않도록 y값에 여유(offset)를 줌
        s=str(txt),            # 표기할 값
        color='blue',       # 글자 색상
        ha='center'          # 정렬: 가운데 정렬
    )

plt.show()
```
